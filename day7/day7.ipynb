
- Five of a kind, where all five cards have the same label: AAAAA
  - `(.)\1\1\1\1` 
- Four of a kind, where four cards have the same label and one card has a different label: AA8AA
  - `(.)\1\1\1.` 
- Full house, where three cards have the same label, and the remaining two cards share a different label: 23332
  - `(.)\1\1(.)\2`
- Three of a kind, where three cards have the same label, and the remaining two cards are each different from any other card in the hand: TTT98
  - `(.)\1\1..` 
- Two pair, where two cards share one label, two other cards share a second label, and the remaining card has a third label: 23432
- One pair, where two cards share one label, and the other three cards have a different label from the pair and each other: A23A4
- High card, where all cards' labels are distinct: 23456


```python

cards = [ x.strip().split(' ') for x in open("d7.b").readlines()]

def score(a):
    l = ['2','3','4','5','6','7','8','9','T','J','Q','K','A']
    return [ l.index(x) for x in list(a) ]
    
def hand_type(a):   

    # count runs of cards
    c = {}
    for j in list(a[0]):
        c[j] = c.get(j, 0) + 1

    # sort by runs of cards
    sorted_dict = {k: v for k, v in sorted(c.items(), key=lambda item: item[1], reverse=True)}

    # figure out the hand type based on the types of runs
    v = list(sorted_dict.values())

    # five of a kind
    if v[0] == 5:
        return [6, score(a[0]),a[0],int(a[1])]
    # four of a kind
    elif v[0] == 4:
        return [5, score(a[0]),a[0],int(a[1])]
    # full house
    elif v[0] == 3 and v[1] == 2:
        return [4, score(a[0]),a[0],int(a[1])]
    # three of a kind
    elif v[0] == 3 and v[1] == 1:
        return [3, score(a[0]),a[0],int(a[1])]
    # two pair
    elif v[0] == 2 and v[1] == 2:
        return [2, score(a[0]),a[0],int(a[1])]
    #one pair    
    elif v[0] == 2 and v[1] == 1:
        return [1, score(a[0]),a[0],int(a[1])]
    # high card
    else:
        return [0, score(a[0]),a[0],int(a[1])]
        return ("high card",a)
    


card_scores = [ hand_type(card) for card in cards ]  
sorted_data = sorted(card_scores, key=lambda x: (x[0], x[1][0], x[1][1], x[1][2], x[1][3], x[1][4]))

total = 0;
for idx,k in enumerate(sorted_data):
    total += k[3] * (idx+1)

total
```
